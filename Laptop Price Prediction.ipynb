```python
import pandas as pd
import numpy as np
import seaborn as sns
df=pd.read_csv('Laptopds.csv')
df[['processor_gnrtn','ram_gb','ssd','hdd']]
```

```python

ram={'0 GB':0,'4 GB GB':4,'8 GB GB':8,'16 GB GB':16,'32 GB':32,'32 GB GB':32,
     '128 GB':128,'256 GB':256,'256 GB GB':256,'512 GB':32,'1024 GB':1024,
     '2048 GB':2048,'3072 GB':3072}

#Converting numerical string data to numerical for calculation
df.ram_gb=[ram[item] for item in df.ram_gb]
df.hdd=[ram[item] for item in df.hdd]
df.ssd=[ram[item] for item in df.ssd]

df['processor_gnrtn'].replace(['4th','7th','8th','9th','10th','11th','12th'],
                              [4,7,8,9,10,11,12],inplace=True)
```

```python
df[['processor_gnrtn','ram_gb','ssd','hdd']]
```

```python
df.describe()
```

```python
#Removing rows with price as 0
df=df[df['old_price']!=0]
df
```

```python
# since only 4 laptops with MediaTek and qualcomm brand are there they wont be helpful in prediction
df = df[ df['processor_brand']!='Qualcomm' ] 
df = df[ df['processor_brand']!='MediaTek' ] 
```

```python
#Dropping latest price column
df.drop('latest_price',axis=1, inplace=True)

#Renaming old_price columns
df.rename(columns={'old_price':'Price'},inplace=True)
df.head(20)

# Since there is no strong correalion between the price and rating, reviews so we drop the column 
df.drop(columns = ['ratings', 'reviews'] , inplace = True)
```

```python
#Fill null values with mean
df.fillna(round(df.mean()),inplace=True)

#Convert processor gen datatype as int
df['processor_gnrtn']=df['processor_gnrtn'].astype(int)
df
```
